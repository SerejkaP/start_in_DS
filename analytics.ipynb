```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salary = pd.read_csv("data/salary.csv")
salary
```

```python
years = [2000+i for i in range(24)]

petroleum_coke = salary.T[0][1:]
rubber_and_plastic = salary.T[1][1:]
fishing = salary.T[2][1:]
```

```python
# Производство кокса и нефтепродуктов
plt.figure(figsize=(15, 7))
plt.plot(years, petroleum_coke)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 5000.0))
plt.title('Производство кокса и нефтепродуктов')
plt.xlabel('Год')
plt.ylabel('Зароботная плата')
plt.show()
```

```python
max_petroleum_coke = max(petroleum_coke)
min_petroleum_coke = min(petroleum_coke)
print("Минимальная за 24 года:", min_petroleum_coke)
print("Максимальная за 24 года:", max_petroleum_coke)
print("Разница max/min:", max_petroleum_coke/min_petroleum_coke)
print("Разница 2024/2000:", petroleum_coke.iloc[23] / petroleum_coke.iloc[0])
```

Для производства кокса и нефтепродуктов наблюдается рост заработной платы до 2017 года. После 2017 года плата снижается и начинает расти только после 2020.

```python
# Производство резиновых и пластмассовых изделий
plt.figure(figsize=(15, 7))
plt.plot(years, rubber_and_plastic)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 5000.0))
plt.title('Производство резиновых и пластмассовых изделий')
plt.xlabel('Год')
plt.ylabel('Зароботная плата')
plt.show()
```

```python
max_rubber_and_plastic = max(rubber_and_plastic)
min_rubber_and_plastic = min(rubber_and_plastic)
print("Минимальная за 24 года:", min_rubber_and_plastic)
print("Максимальная за 24 года:", max_rubber_and_plastic)
print("Разница max/min:", max_rubber_and_plastic/min_rubber_and_plastic)
print("Разница 2024/2000:", rubber_and_plastic.iloc[23] / rubber_and_plastic.iloc[0])
```

Для производства резиновых и пластмассовых изделий рост заработной платы продолжается все года.

```python
# Рыболовство, рыбоводство
plt.figure(figsize=(15, 7))
plt.plot(years, fishing)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 5000.0))
plt.title('Рыболовство, рыбоводство')
plt.xlabel('Год')
plt.ylabel('Зароботная плата')
plt.show()
```

Для рыболовства, рыбоводства заработная плата активно растет на протяжении всех лет. После 2014 года и 2021 года заработная плата начинает расти более активно.

```python
# На одном графике
plt.figure(figsize=(15, 7))
plt.plot(years, petroleum_coke, color='r', label='Производство кокса и нефтепродуктов')
plt.plot(years, rubber_and_plastic, color='b', label='Производство резиновых и пластмассовых изделий')
plt.plot(years, fishing, color='g', label='Рыболовство, рыбоводство')
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 5000.0))
plt.legend()
plt.title('Все деятельности')
plt.xlabel('Год')
plt.ylabel('Зароботная плата')
plt.show()
```

```python
inflation_data = pd.read_csv("data/inflation.csv")
inflation = inflation_data[['Год', 'Всего']]
inf_years = inflation_data['Год'][1:25]
inf_total = inflation_data['Всего'][1:25]
inflation.T
```

```python
plt.figure(figsize=(15, 6))
plt.plot(inf_years, inf_total)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.title('График инфляции')
plt.xlabel('Год')
plt.ylabel('Инфляция')
plt.show()
```

```python
def take_real_salary(year_salary, initial_year, finish_year):
  real = year_salary
  for year in range(finish_year, initial_year, -1):
    real /= 1.0 + inflation[inflation['Год'] == year].iloc[0]['Всего'] / 100
  return real

real_petroleum_coke = [salary['2000'][0]]
real_rubber_and_plastic = [salary['2000'][1]]
real_fishing = [salary['2000'][2]]

for i in range(2001, 2024):
  year_salary = salary[str(i)]

  real_petroleum_coke.append(take_real_salary(year_salary[0], 2000, i))
  real_rubber_and_plastic.append(take_real_salary(year_salary[1], 2000, i))
  real_fishing.append(take_real_salary(year_salary[2], 2000, i))

  prev_salary = year_salary

```

```python
# Производство кокса и нефтепродуктов
plt.figure(figsize=(15, 7))
plt.plot(years, real_petroleum_coke, color='b', label='С учетом инфляции по ценам 2000г')
plt.plot(years, petroleum_coke, color='r', label='Без учета инфляции')
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 5000.0))
plt.title('Производство кокса и нефтепродуктов')
plt.legend()
plt.xlabel('Год')
plt.ylabel('Зароботная плата')
plt.show()
```

```python
max_real_petroleum_coke = max(real_petroleum_coke)
min_real_petroleum_coke = min(real_petroleum_coke)
print("Минимальная за 24 года:", min_real_petroleum_coke)
print("Максимальная за 24 года:", max_real_petroleum_coke)
print("Разница max/min:", max_real_petroleum_coke/min_real_petroleum_coke)
print("Разница 2024/2000:", real_petroleum_coke[23] / real_petroleum_coke[0])
```

Для производства кокса и нефтепродуктов рост заработной платы с учетом инфляции длился до 2017 года с небольшой просадкой в 2014-2015 году. После 2017 года заработная плата уменьшается и начинает расти только в 2022 году, не превышая максимальной.

```python
# Производство резиновых и пластмассовых изделий
plt.figure(figsize=(15, 7))
plt.plot(years, real_rubber_and_plastic, color='b', label='С учетом инфляции по ценам 2000г')
plt.plot(years, rubber_and_plastic, color='r', label='Без учета инфляции')
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 5000.0))
plt.title('Производство резиновых и пластмассовых изделий')
plt.xlabel('Год')
plt.legend()
plt.ylabel('Зароботная плата')
plt.show()
```

```python
max_real_rubber_and_plastic = max(real_rubber_and_plastic)
min_real_rubber_and_plastic = min(real_rubber_and_plastic)
print("Минимальная за 24 года:", min_real_rubber_and_plastic)
print("Максимальная за 24 года:", max_real_rubber_and_plastic)
print("Разница max/min:", max_real_rubber_and_plastic/min_real_rubber_and_plastic)
print("Разница 2024/2000:", real_rubber_and_plastic[23] / real_rubber_and_plastic[0])
```

Для производства резиновых и пластмассовых изделий заработная плата растет менее значительно с небольшой просадкой в 2013-2015гг.
